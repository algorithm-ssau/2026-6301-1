```python
# Настройка VK токена (хранение)

import json

# Создаем файл с конфигурацией (не загружайте в GitHub!)
config_path = os.path.join(BOT_FOLDER, 'config.json')

# Первый запуск - введите токен
if not os.path.exists(config_path):
    VK_TOKEN = input("🔑 Введите ваш VK API токен: ")

    config = {
        'vk_token': VK_TOKEN,
        'bot_version': '1.0.0'
    }

    with open(config_path, 'w') as f:
        json.dump(config, f)
    print("Конфигурация сохранена!")
else:
    with open(config_path, 'r') as f:
        config = json.load(f)
        VK_TOKEN = config['vk_token']
    print("Конфигурация загружена")

# Проверяем токен
print(f"Бот будет использовать токен: {VK_TOKEN[:10]}...")
```
